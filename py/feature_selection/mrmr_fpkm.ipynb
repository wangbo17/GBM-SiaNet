```python
import pandas as pd
from mrmr import mrmr_classif
```

```python
train = pd.read_csv('inputs/fpkm_train.csv', index_col = 0)
label_train = pd.read_csv('inputs/label_train.csv')
X_train, y_train = train.values, label_train['label'].values

X_train = pd.DataFrame(X_train, columns=train.columns.values)
y_train = pd.Series(y_train)
```

```python
selected_features = mrmr_classif(X=X_train, y=y_train, K=500)
print("Selected features:", selected_features)
```

```python
features_mrmr = pd.DataFrame({
    'Feature': selected_features,
})
features_mrmr.to_csv('inputs/mrmr_features_fpkm.csv', index=False)
```
